### 1. Заповніть пропущені значення в датасеті з файлу **'train.csv'**.
* Обґрунтуйте вибір способу заповнення пропущених значень.

```python
import pandas as pd


df_data = pd.read_csv("train.csv")
df_data
```

```python
print('All columns of dataset:')
df_data.columns
```

```python
print('Columns with missing values:')
df_data.isnull().sum()
```

```python
print('Number of all missing values:')
df_data.isnull().sum().sum()
```

```python
print('Missing values of the Age column Before filling:' )
df_data['age'].isnull().sum()
```

### Visualisation of the Age column Before filling with the new values:

```python
import matplotlib.pyplot as plt


plt.hist(df_data['age'], color='purple', edgecolor='red', bins=10)
plt.title('BEFORE filling new data')
plt.xlabel('Distribution of passengers by age')
plt.show()
```

### Filling a missing data with average values (Mean method):

```python
df_data['age'].fillna(df_data['age'].mean(), inplace=True)
df_data
```

```python
print('Missing values of Age column After filling:' )
df_data['age'].isnull().sum()
```

### Why I chose the Mean method:
* **In this case I filled missing data with the mean values. To my mind it is a good approach
of imputing missing data of numeric columns. Also, I will know the average age of all Titanic's passengers. The average age may be necessary for further data analysis**.
